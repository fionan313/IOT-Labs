```python
fig, ax = plt.subplots()
ax.scatter(data['numBedrooms'], data['price'])
ax.set_xlabel('Number of bedrooms')
ax.set_ylabel('Price')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data['numBedrooms'], data['price'], alpha=0.5)
ax.set_xlabel('Number of Bedrooms')
ax.set_ylabel('Price (€)')
ax.set_title('Price vs. Number of Bedrooms')
plt.tight_layout()
plt.show()
```

```python
data['publishDate'] = pd.to_datetime(data['publishDate'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(data['publishDate'], data['price'], alpha=0.5)
ax.set_xlabel('Publish Date')
ax.set_ylabel('Price (€)')
ax.set_title('Price vs. Publish Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data['numBathrooms'], data['price'], alpha=0.5)
ax.set_xlabel('Number of Bathrooms')
ax.set_ylabel('Price (€)')
ax.set_title('Price vs. Number of Bathrooms')
plt.tight_layout()
plt.show()
```

```python
data_clean = data.dropna(subset=['propertySize', 'price'])

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data_clean['propertySize'], data_clean['price'], alpha=0.5)
ax.set_xlabel('Property Size (m²)')
ax.set_ylabel('Price (€)')
ax.set_title('Price vs. Property Size')
plt.tight_layout()
plt.show()
```

```python
data['county'] = data['title'].str.extract(r'Co\.\s*([A-Za-z]+)')

# average price per county
avg_price_county = data.groupby('county')['price'].mean().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(12, 6))
avg_price_county.plot(kind='bar', ax=ax)
ax.set_xlabel('County')
ax.set_ylabel('Average Price (€)')
ax.set_title('Average House Price by County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_price_type = data.groupby('propertyType')['price'].mean().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
avg_price_type.plot(kind='bar', ax=ax, color='steelblue')
ax.set_xlabel('Property Type')
ax.set_ylabel('Average Price (€)')
ax.set_title('Average House Price by Property Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
correlation = data.corr(numeric_only=True)
```

```python
correlation['price']['numBedrooms']
```

```python
correlation['price']
```

```python
correlation = correlation.round(decimals=2)
names = list(correlation.columns)
corr = correlation.to_numpy()
```

```python
fig, ax = plt.subplots(figsize=(14, 12))

# heatmap
im = ax.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1, aspect='auto')

# labels
ax.set_xticks(np.arange(len(names)))
ax.set_yticks(np.arange(len(names)))
ax.set_xticklabels(names, fontsize=10)
ax.set_yticklabels(names, fontsize=10)

# Rotate x-axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

cbar = plt.colorbar(im, ax=ax, shrink=0.8)
cbar.set_label('Correlation Coefficient', rotation=270, labelpad=25, fontsize=12)

# Add correlation values
for i in range(len(names)):
    for j in range(len(names)):

        text_color = "white" if abs(corr[i, j]) > 0.5 else "black"
        text = ax.text(j, i, f'{corr[i, j]:.2f}',
                      ha="center", va="center", color=text_color, 
                      fontsize=9, weight='bold' if abs(corr[i, j]) > 0.7 else 'normal')

# title
ax.set_title('Correlation Matrix: Irish Property Data', 
             pad=20, fontsize=16, weight='bold')

#  grid lines
ax.set_xticks(np.arange(len(names))-.5, minor=True)
ax.set_yticks(np.arange(len(names))-.5, minor=True)
ax.grid(which="minor", color="gray", linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()
```
